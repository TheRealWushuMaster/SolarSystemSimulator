```python
from skyfield.api import load
import matplotlib.pyplot as plt

# Load the JPL planetary ephemeris data
eph = load('de421.bsp')

# Get the position of the Sun and Earth at a specific time
ts = load.timescale()
t = ts.utc(2023, 1, 1)
sun = eph['sun']
earth = eph['earth']
position_sun = sun.at(t).position.km
position_earth = earth.at(t).position.km

# Plot the positions
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(position_sun[0], position_sun[1], position_sun[2], c='yellow', label='Sun')
ax.scatter(position_earth[0], position_earth[1], position_earth[2], c='blue', label='Earth')

# Set plot attributes
ax.set_xlabel('X (km)')
ax.set_ylabel('Y (km)')
ax.set_zlabel('Z (km)')
ax.set_title('Solar System')
ax.legend()

# Show the plot
plt.show()
```

```python
from skyfield.api import load
import matplotlib.pyplot as plt

# Load the JPL planetary ephemeris data
eph = load('de421.bsp')

# Get the position of the Sun and Earth at a specific time
ts = load.timescale()
t = ts.utc(2023, 1, 1)
sun = eph['sun']
earth = eph['earth']
position_sun = sun.at(t).position.au
position_earth = earth.at(t).position.au

# Plot the positions
fig, ax = plt.subplots()
ax.set_aspect('equal')

# Plot the ecliptic plane
ax.plot([-1.5, 1.5], [0, 0], color='gray', linestyle='--')

# Plot the Sun and Earth
ax.scatter(position_sun[0], position_sun[1], c='yellow', label='Sun')
ax.scatter(position_earth[0], position_earth[1], c='blue', label='Earth')

# Set plot attributes
ax.set_xlabel('X (AU)')
ax.set_ylabel('Y (AU)')
ax.set_title('Solar System - Ecliptic Plane')
ax.legend()

# Show the plot
plt.show()
```

```python
import pygame
from skyfield.api import load

# Constants for the window size
WIDTH, HEIGHT = 800, 600

# Initialize PyGame
pygame.init()
screen = pygame.display.set_mode((WIDTH, HEIGHT))
clock = pygame.time.Clock()

# Load the JPL planetary ephemeris data
eph = load('de421.bsp')

# Get the position of the Sun and Earth at a specific time
ts = load.timescale()
t = ts.utc(2023, 1, 1)
sun = eph['sun']
earth = eph['earth']
position_sun = sun.at(t).position.au
position_earth = earth.at(t).position.au

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Clear the screen
    screen.fill((0, 0, 0))

    # Convert the positions to screen coordinates
    x_sun = int((position_sun[0] + 1) * WIDTH / 2)
    y_sun = int((position_sun[1] + 1) * HEIGHT / 2)
    x_earth = int((position_earth[0] + 1) * WIDTH / 2)
    y_earth = int((position_earth[1] + 1) * HEIGHT / 2)

    # Draw the Sun and Earth
    pygame.draw.circle(screen, (255, 255, 0), (x_sun, y_sun), 10)
    pygame.draw.circle(screen, (0, 0, 255), (x_earth, y_earth), 5)

    # Update the display
    pygame.display.flip()
    clock.tick(60)

# Quit PyGame
pygame.quit()
```

```python
import pygame
from skyfield.api import load

# Constants for the window size and sidebar width
WIDTH, HEIGHT = 800, 600
SIDEBAR_WIDTH = 200

# Initialize PyGame
pygame.init()
screen = pygame.display.set_mode((WIDTH + SIDEBAR_WIDTH, HEIGHT))
clock = pygame.time.Clock()

# Load the JPL planetary ephemeris data
eph = load('de421.bsp')

# Define colors
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)

# Define the sidebar dimensions
sidebar_rect = pygame.Rect(WIDTH, 0, SIDEBAR_WIDTH, HEIGHT)

# Define the buttons
button_rect = pygame.Rect(WIDTH + 20, 20, SIDEBAR_WIDTH - 40, 50)

# Get the position of the Sun and Earth at a specific time
ts = load.timescale()
t = ts.utc(2023, 1, 1)
sun = eph['sun']
earth = eph['earth']
position_sun = sun.at(t).position.au
position_earth = earth.at(t).position.au

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Clear the screen
    screen.fill(BLACK)

    # Draw the Solar System representation
    pygame.draw.circle(screen, WHITE, (WIDTH // 2, HEIGHT // 2), 50)  # Example: Sun
    pygame.draw.circle(screen, WHITE, (WIDTH // 2 + 100, HEIGHT // 2), 10)  # Example: Earth

    # Draw the sidebar
    pygame.draw.rect(screen, WHITE, sidebar_rect)

    # Draw the buttons
    pygame.draw.rect(screen, WHITE, button_rect)
    button_text = pygame.font.SysFont(None, 30).render("Button", True, BLACK)
    screen.blit(button_text, (WIDTH + 50, 30))

    # Update the display
    pygame.display.flip()
    clock.tick(60)

# Quit PyGame
pygame.quit()
```

```python
import pygame
from skyfield.api import load

# Constants for the window size and sidebar width
WIDTH, HEIGHT = 800, 600
SIDEBAR_WIDTH = 200

# Initialize PyGame
pygame.init()
screen = pygame.display.set_mode((WIDTH + SIDEBAR_WIDTH, HEIGHT))
clock = pygame.time.Clock()

# Load the JPL planetary ephemeris data
eph = load('de421.bsp')

# Load textures for celestial bodies
sun_texture = pygame.image.load("oct_1_2015_flare.jpg")  # Example: Sun texture image
earth_texture = pygame.image.load("as10-34-5013.jpg")  # Example: Earth texture image

# Get the position of the Sun and Earth at a specific time
ts = load.timescale()
t = ts.utc(2023, 1, 1)
sun = eph['sun']
earth = eph['earth']
position_sun = sun.at(t).position.au
position_earth = earth.at(t).position.au

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Clear the screen
    screen.fill((0, 0, 0))

    # Draw the Solar System representation
    sun_rect = sun_texture.get_rect(center=(WIDTH // 2, HEIGHT // 2))
    screen.blit(sun_texture, sun_rect)

    earth_rect = earth_texture.get_rect(center=(WIDTH // 2 + 100, HEIGHT // 2))
    screen.blit(earth_texture, earth_rect)

    # Draw the sidebar
    pygame.draw.rect(screen, (255, 255, 255), (WIDTH, 0, SIDEBAR_WIDTH, HEIGHT))

    # Update the display
    pygame.display.flip()
    clock.tick(60)

# Quit PyGame
pygame.quit()
```
